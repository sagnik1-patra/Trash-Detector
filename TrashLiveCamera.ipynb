```python
# ‚úÖ Install required libraries (if not installed)
!pip install tensorflow opencv-python

# ‚úÖ Imports
import cv2
import numpy as np
import tensorflow as tf

# ‚úÖ Load the trained model
model_path = r"C:\Users\sagni\Downloads\Trash-Detector\waste_classifier.h5"
model = tf.keras.models.load_model(model_path)

# ‚úÖ Class labels (same as your dataset folders)
class_labels = ['cardboard', 'glass', 'metal', 'paper', 'plastic', 'trash']

# ‚úÖ Preprocessing function
def preprocess_frame(frame):
    img = cv2.resize(frame, (224, 224))
    img = img.astype('float32') / 255.0
    return np.expand_dims(img, axis=0)

# ‚úÖ Start webcam
cap = cv2.VideoCapture(0)  # 0 = default camera

if not cap.isOpened():
    print("‚ùå Cannot access camera")
    exit()

print("‚úÖ Camera opened. Press 'q' to quit.")

while True:
    ret, frame = cap.read()
    if not ret:
        print("‚ùå Failed to grab frame")
        break

    # Preprocess and predict
    input_img = preprocess_frame(frame)
    prediction = model.predict(input_img)
    class_index = np.argmax(prediction)
    confidence = np.max(prediction)

    label_text = f"{class_labels[class_index]} ({confidence*100:.2f}%)"

    # Draw label on frame
    cv2.putText(frame, label_text, (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    # Display the resulting frame
    cv2.imshow('Live Trash Detector', frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release camera and close windows
cap.release()
cv2.destroyAllWindows()
print("üõë Camera closed.")
```
